```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students.csv")
df
```

```python
df["Average"] = (df["Math"] + df["Science"] + df["English"]) / 3
df
```

```python
top_student = df.loc[df["Average"].idxmax()]
print("Top Student:")
print(top_student)
```

```python
print("Highest Average:", df["Average"].max())
print("Lowest Average:", df["Average"].min())
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Average"])
plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Average"])
plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")

plt.savefig("marks_chart.png")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    df["Average"],
    labels=df["Name"],
    autopct="%1.1f%%"
)
plt.title("Student Performance Distribution")
plt.show()
```

```python
print("Total Students:", len(df))
print("Average Class Score:", round(df["Average"].mean(), 2))
print("Top Student:", top_student["Name"])
```
